```python
import pandas as pd
import numpy as np
import tensorflow as tf
import scikeras
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasRegressor
from sklearn.pipeline import Pipeline
```

```python
music = pd.read_csv("MusicDataset.csv")
music.head()
```

```python
music = music.drop(['Unnamed: 0', 'track_id', 'artists', 'album_name', 'track_name','explicit', 'loudness', 'mode', 
       'track_genre'], axis = 1)
music.describe
```

```python
x = music.drop(columns = 'valence')
x
```

```python
y = music['valence']
y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.25, random_state = 88)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## ANN attempt 1

```python
model = Sequential()
 
model.add(Dense(units=5, input_dim=11, kernel_initializer='normal', activation='relu'))
model.add(Dense(units=5, kernel_initializer='normal', activation='tanh'))
model.add(Dense(1, kernel_initializer='normal'))
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(X_train, Y_train ,batch_size = 20, epochs = 15, verbose=1)
```

```python
score = model.evaluate(x, y, verbose=1)
score
```

```python
y_pred = model.predict(x)
l = len(y)
acc = sum([np.round(y_pred[i])==y[i] for i in range(l)])/l
print("The accuracy of the model is", acc)
```

## ANN attempt 2

```python
model2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(12, activation = 'relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(4, activation = 'softmax')
])
```
